### Matplotlib example workflow

```python
# 0. Import matplotlib and get ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x=[1,2,3,4]
y=[11,22,33,44]

#2. Setup data
fig,ax=plt.subplots(figsize=(10,10)) # (height,width)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("data/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows=2, 
                                           ncols=2,
                                           figsize=(10,5))

#plot to each different axis
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig,ax= plt.subplots(nrows=2, 
                    ncols=2,
                    figsize=(10,5))

#plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

### Which one Should you use? (pyplot vs matplotlib Object Oriented Methor or the 00 Method?)

* When plotting somethind quickly, okay to use the pyplot method
* When plotting something more advanced, use OO method

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target');
# c = Colour bar
```

```python
# OO method with pyplot method
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

# ax is refering to th are where the plot is made

# This change the x axis and sets it from 45 to 100
# ax.set_xlim([45,100]) 
```

```python
# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# This legend_elements() looks up in the plot table and searches for the c label

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
# Subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                        ncols=1,
                        figsize=(10,10),sharex=True)

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       ylabel="Cholestrol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax0
ax0.axhline(over_50["chol"].mean(),linestyle='--');

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax1
ax1.axhline(over_50["thalach"].mean(),linestyle='--');
```

## Customizing Matplotlib plots and getting stylish

```python
# See the different styles avilable
plt.style.available
```

```python
# Default
car_sales["Price"].plot();
```

```python
plt.style.use('ggplot')
car_sales["Price"].plot()
```

```python
# Creating a data
x=np.random.randn(10,4)
df=pd.DataFrame(x,columns=['a','b','c','d'])
ax=df.plot(kind='bar');
```

```python
# Customize our plot with the set() method
ax=df.plot(kind="bar")

# Add some labels and a title
ax.set(title="Random Number Bar Graph from DataFrame",
        xlabel="Row number",
        ylabel="Random number")

# Make the legend visible
ax.legend().set_visible(True)
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

# OO method from scratch
fig,ax=plt.subplots(figsize=(10,6))

#plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter") # This changes the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Add a legend
ax.legend(*scatter.legend_elements(),title="Target")
# This legend_elements() looks up in the plot table and searches for the c label

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),linestyle='--');
```

```python
# Customizing the x and y axis limitations

plt.style.use("ggplot")

# Subplot of chol, age, thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                        ncols=1,
                        figsize=(10,10),sharex=True)

# Add data to ax0
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                   cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholestrol Levels",
       xlabel="Age",
       ylabel="Cholestrol")

# Change the x limitation
ax0.set_xlim([50,80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax0
ax0.axhline(over_50["chol"].mean(),linestyle='-.');

# Add data to ax1
scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="winter")

# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate Levels",
       xlabel="Age",
       ylabel="Max Heart Rate");

# Change the x limitation
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(),title="Target")

# Add a horizontal line to ax1
ax1.axhline(over_50["thalach"].mean(),linestyle='-.')

# Add a title toi the entire figure
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight='bold');
```
